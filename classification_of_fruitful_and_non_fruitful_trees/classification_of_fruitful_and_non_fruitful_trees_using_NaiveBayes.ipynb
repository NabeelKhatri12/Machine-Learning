```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
data = pd.read_csv('C:/Users/Dua/Desktop/classification_of_trees/training.csv')
```

```python
data
```

```python
data.drop(data.tail(1).index,inplace=True)
```

```python
data
```

```python
# Number of fruitful trees
data[data.fruitful_or_non_fruitful == 1].size
```

```python
# Number of non_fruitful trees
data[data.fruitful_or_non_fruitful == 0].size
```

```python
data.size
```

```python
dups_values = data.pivot_table(index=['fruitful_or_non_fruitful'], aggfunc='size')
print (dups_values)
```

```python
names = data['names']
```

```python
cv = CountVectorizer()
x = cv.fit_transform(names)
```

```python
cv.get_feature_names()
```

```python
# For training and testing
from sklearn.model_selection import train_test_split
```

```python
# Features
x
# Labels
y = data.fruitful_or_non_fruitful
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 42)
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(x_train, y_train)
clf.score(x_test, y_test)
```

```python
# Accuracy of model
print("Accuracy of our Model = ", clf.score(x_test, y_test) * 100, "%")
```

```python
# Accuracy of model
print("Accuracy of our Model = ", clf.score(x_train, y_train) * 100, "%")
```
