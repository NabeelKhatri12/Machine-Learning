```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
#from sklearn.feature_extraction import DictVectorizer
```

```python
data = pd.read_csv('C:/Users/Dua/Desktop/ML Coding/DataSet_Gender.csv')
```

```python
# Number of males name
data[data.gender == 'm'].size
```

```python
data.gender.replace({'f': 0, 'm': 1}, inplace = True)
```

```python
data
```

```python
data.gender.unique()
```

```python
names = data['name']
```

```python
cv = CountVectorizer()
x = cv.fit_transform(names)
```

```python
cv.get_feature_names()
```

```python
# For training and testing
from sklearn.model_selection import train_test_split
```

```python
# Features
x
# Labels
y = data.gender
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 42)
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(x_train, y_train)
clf.score(x_test, y_test)
```

```python
# Accuracy of model
print("Accuracy of our Model = ", clf.score(x_test, y_test) * 100, "%")
```

```python
# Accuracy of model
print("Accuracy of our Model = ", clf.score(x_train, y_train) * 100, "%")
```
